```python
import random

contbaixa = 0        #  Declarando as variaveis e armazenando 0 nelas
contadequada = 0
contalta = 0
def classificar(valor):  # Função Criada para classificar o valor recebido em BAixa, Adequada ou Alta
    if valor < 3:
        return "Baixa"
    elif valor <= 7:
        return "Adequada"
    else:
        return "Alta"
for i in range(1, 11):  #Utilizei o for como estrutura de Repetição pois sabia a qtd de leituras que são 10
    concentracao = round(random.uniform(0, 10), 1)  # Gera o valor aleatório e arredonda para 1 casa decimal
    classe = classificar(concentracao)  # Chama a função para saber qual é a classe
    if classe == "Baixa":     # O if Funciona como um contador para armazenar a qtd de vezes que a concentração foi baixa alta ou adequada
        contbaixa = contbaixa + 1
    elif classe == "Adequada":
        contadequada = contadequada + 1
    else:
        contalta = contalta + 1
    print(f"Leitura {i}: Concentração = {concentracao} mol/L -> {classe}")   # Como o print esta dentro do for ele vai ser mostrado 10 vezes o LEitura{i} mostra qual e a leitura do momento 1,2,3....

predominante = "Baixa"    # Aqui eu dei um chute de quem seria o Predominante pois meu programa nao sabe quem e ainda. Nessa linha de copdigo o programa apenas rodou o for
maiorquantidade = contbaixa
if contadequada > maiorquantidade:  # Aqui que eu realmente vou verificar quem e o predominante
    predominante = "Adequada"
    maiorquantidade = contadequada
if contalta > maiorquantidade:
    predominante = "Alta"


print("\nResumo:")
print(f"Baixa: {contbaixa} ocorrências")
print(f"Adequada: {contadequada} ocorrências")
print(f"Alta: {contalta} ocorrências")
print(f"Condição predominante: {predominante}")
```
